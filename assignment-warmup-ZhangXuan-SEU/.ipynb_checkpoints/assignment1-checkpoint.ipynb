```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import warnings
import time
warnings.filterwarnings("ignore")
```

```python
def loadData():
    with np.load('notMNIST.npz') as data :
        Data, Target = data ['images'], data['labels']
        posClass = 2
        negClass = 9
        dataIndx = (Target==posClass) + (Target==negClass)
        Data = Data[dataIndx]/255.
        Target = Target[dataIndx].reshape(-1, 1)
        Target[Target==posClass] = 1
        Target[Target==negClass] = 0
        np.random.seed(421)
        randIndx = np.arange(len(Data))
        np.random.shuffle(randIndx)
        Data, Target = Data[randIndx], Target[randIndx]
        trainData, trainTarget = Data[:3500], Target[:3500]
        validData, validTarget = Data[3500:3600], Target[3500:3600]
        testData, testTarget = Data[3600:], Target[3600:]
    return trainData, validData, testData, trainTarget, validTarget, testTarget
```

```python
trainData, validData, testData, trainTarget, validTarget, testTarget=loadData()
```

```python
print("Training set", trainData .shape)
print("Validation set", validData.shape)
print("Test set", testData.shape)
```

```python
validData=validData.reshape((100,28*28))
testData=testData.reshape((145,28*28))
trainData=trainData.reshape((3500,28*28))
```

# 1.Linear Regression

## 1.1 Loss Function and Gradient

```python
def MSE(W, b, x, y, reg):

    N=len(x)
    predictions=x.dot(W)+b
    prediction_classes=(predictions>=0.5).astype(int)
    mse=np.sum(np.square(predictions-y))/(2*N)+(1/2*reg)*np.sum(np.square(W))
    
    return mse,prediction_classes
        
```

```python
def gradMSE(W, b, x, y, reg):
    N=len(x)
    predictions=x.dot(W)+b

    W_grad=(1/N)*(x.T.dot(predictions-y))+reg*W
    b_grad=(1/N)*(np.sum(predictions-y))
    
    return W_grad, b_grad
```

## 1.2 Gradient Descent Implementation

```python
def grad_descent(W, b, x, y, alpha, epochs, reg,title,error_tol=1e-7):
    cost=[]
    cost_valid=[]
    cost_test=[]
    
    accuracy=[]
    accuracy_valid=[]
    accuracy_test=[]
    
    N=len(x)
    start_time=time.time()
    
    for i in range(epochs+1):
        W_grad=gradMSE(W, b, x, y, reg)[0]
        b_grad=gradMSE(W, b, x, y, reg)[1]

        W-=alpha*W_grad
        b-=alpha*b_grad

        eps=np.sum(abs(alpha*W_grad))/N
        
        Current_time=time.time()-start_time
        cost.append(MSE(W, b, x, y, reg)[0])
        cost_valid.append(MSE(W, b, validData, validTarget, reg)[0])
        cost_test.append(MSE(W, b, testData, testTarget, reg)[0])

        accuracy.append(np.mean((MSE(W, b, x, y, reg)[1]==y)))
        accuracy_valid.append(np.mean(MSE(W, b, validData, validTarget, reg)[1]==validTarget))
        accuracy_test.append(np.mean(MSE(W, b, testData,testTarget, reg)[1]==testTarget))
        if i%500==0 and i!=0:
            print('Epochs:%d - Trainloss:%.4f - Trainaccuracy:%.4f'%(i,cost[i],accuracy[i]))
            print('           - Validloss:%.4f - Validaccuracy:%.4f'%(cost_valid[i],accuracy_valid[i]))
            print('           - Testloss:%.4f - Testaccuracy:%.4f'%(cost_test[i],accuracy_test[i]))
            print('           - Time:%.4f'%(Current_time))
        if eps<error_tol:
            print('Finished when epochs=%i'%i)
            print('Epochs:%d - Trainloss:%.4f - Trainaccuracy:%.4f'%(i,cost[i],accuracy[i]))
            print('           - Validloss:%.4f - Validaccuracy:%.4f'%(cost_valid[i],accuracy_valid[i]))
            print('           - Testloss:%.4f - Testaccuracy:%.4f'%(cost_test[i],accuracy_test[i]))
            print('           - Time:%.4f'%(Current_time))
            break

    plt.figure(figsize=(12,8))
    plt.plot(cost,color='b',label='training loss')
    plt.plot(cost_valid,color='r',label='validation loss')
    plt.plot(cost_test,color='g',label='test loss')
    plt.ylabel('loss',fontsize=16)
    plt.xlabel('iteration',fontsize=18)
    plt.title(title+' loss /reg=%1.2f'%reg+' /learning rate=%1.4f'%alpha,fontsize=20)
    plt.legend(loc='upper right')
            
    plt.figure(figsize=(12,8))
    plt.plot(accuracy,color='b',label='training accuracy')
    plt.plot(accuracy_valid,color='r',label='validation accuracy')
    plt.plot(accuracy_test,color='g',label='test accuracy')
    plt.ylabel('accuracy',fontsize=16)
    plt.xlabel('iteration',fontsize=18)
    plt.title(title+' accuracy  /reg=%1.2f'%reg+' /learning rate=%1.4f'%alpha,fontsize=20)
    plt.legend(loc='lower right')

    return W,b,cost,cost_valid,cost_test,eps
            
        
```

```python
W1=np.zeros((784,1))
b1=np.zeros([])
x=trainData
y=trainTarget
alpha=0.005
epochs=5000
reg=0
title='linear regression'
weight,bias,trainloss,validloss,testloss,error_total=grad_descent(W1,b1,x,y,alpha,epochs,reg,title)
```

## 1.5 Comparing Batch GD with normal equation

```python
def normal_equation(x,y):
    start_time=time.time()
    W1=np.linalg.inv(np.dot(x.T,x))
    W2=np.dot(W1,x.T)
    W=np.dot(W2,y)
    end_time=time.time()
    training_time=end_time-start_time
    return W,training_time
    
```

```python
x=trainData
y=trainTarget
W,training_time=normal_equation(x,y)
```

```python
trainloss_nor=MSE(W, 0, x, y, 0)[0]
trainacc_nor=np.mean(MSE(W,0,x,y,0)[1]==y)
print("training loss of normal equation =%.3f"%trainloss_nor)
print("training accuracy of normal equation =%.3f"%trainacc_nor)
print ("training time of normal equation=%.3f"%training_time)
```

# 2. logistic regression

## 2.1 Loss Function and Gradient

```python
def crossEntropyLoss(W, b, x, y, reg):
    z=x.dot(W)+b
    predictions=1./(1.+np.exp(-z))
    prediction_classes=(predictions>=0.5).astype(int)
    N=len(x)
    ce=(1/N)*(np.sum((-1)*y*np.log(predictions)-(1-y)*np.log(1-predictions)))+(reg/2)*np.sum(np.square(W))
    return ce,prediction_classes
```

```python
def gradCE(W, b, x, y, reg):
    N=len(x)
    z=x.dot(W)+b
    predictions=1/(1+np.exp(-z))
    W_grad=(1/N)*(x.T.dot(predictions-y))+reg*W
    b_grad=(1/N)*(np.sum(predictions-y))
    return W_grad,b_grad
```

## 2.2 Learning

```python
def grad_descent_edited(W, b, x, y, alpha, epochs, reg, EPS=1e-7, lossType="None"):
    cost=[]
  
    accuracy=[]
    
    N=len(x)
    
    if lossType=='MSE':
        print('Linear Regression')
        for i in range(epochs+1):
            W_grad=gradMSE(W, b, x, y, reg)[0]
            b_grad=gradMSE(W, b, x, y, reg)[1]
            
            W-=alpha*W_grad
            b-=alpha*b_grad
            
            eps=np.sum(abs(alpha*W_grad))/N

            cost.append(MSE(W, b, x, y, reg)[0])

            accuracy.append(np.mean((MSE(W, b, x, y, reg)[1]==y)))
            
            if i%500==0 :
                print('Epochs:%d - Trainloss:%.4f - Trainaccuracy:%.4f'%(i,cost[i],accuracy[i]))
            if eps<EPS:
                print('Finished when epochs=%i'%(i+1))
                break
       
    elif lossType=='CE':
        print('Logistic Regression')
        for i in range(epochs+1):
            W_grad=gradCE(W, b, x, y, reg)[0]
            b_grad=gradCE(W, b, x, y, reg)[1]
            
            W-=alpha*W_grad
            b-=alpha*b_grad
            
            eps=np.sum(abs(alpha*W_grad))/N
            
            
            cost.append(crossEntropyLoss(W, b, x, y, reg)[0])
            
            accuracy.append(np.mean(crossEntropyLoss(W, b, x, y, reg)[1]==y))
            if i%500==0 :
                print('Epochs:%d - loss:%.4f - accuracy:%.4f'%(i,cost[i],accuracy[i]))
                
            if eps<EPS:
                print('Finished when epochs=%i'%(i+1))
                break    
    
    
    
    return W,b,cost,accuracy,eps
        
```

```python
def plot(cost,accuracy,title,reg,alpha):
    epochs=range(500,3500)
    plt.figure(figsize=(12,8))
    plt.plot(epochs,cost[500:],color='b')
    plt.ylabel('loss',fontsize=16)
    plt.xlabel('iteration',fontsize=18)
    plt.title(title+' training loss /reg=%1.2f'%reg+' /learning rate=%1.3f'%alpha,fontsize=20)
    plt.savefig('loss.png')
    plt.show()
            
    plt.figure(figsize=(12,8))
    plt.plot(epochs,accuracy[500:],color='b')
    plt.ylabel('accuracy',fontsize=16)
    plt.xlabel('iteration',fontsize=18)
    plt.title(title+' training accuracy  /reg=%1.2f'%reg+' /learning rate=%1.3f'%alpha,fontsize=20)
    plt.savefig('accuracy.png')

    plt.show()
   
```

# 3. SGD

## 3.1 Building the Computational Graph

```python
def buildGraph(alpha,beta1,beta2,epsilon,optimizertype,loss=None):
    
    # Create tensors for training data and labels 
    x = tf.placeholder(tf.float32, shape=(None, 784),name='x')
    y=tf.placeholder(tf.float32,shape=(None,1),name='y')


    #for regula
    reg=tf.placeholder(tf.float32,name='reg')

    seed=tf.set_random_seed(421)

    #for weight and bias tensors
    weight=tf.Variable(tf.truncated_normal(shape=(784,1),stddev=0.5,seed=seed))
    bias = tf.Variable(tf.zeros([1]))
    
    # Training computation.
    logits = tf.matmul(x, weight) + bias 
    
    # L2 loss
    reg_loss = tf.nn.l2_loss(weight)

    if loss == "MSE":
        # Regular loss
        loss=tf.reduce_mean(tf.square(logits-y))/2
        
        loss_total=loss+reg*reg_loss
        
        if optimizertype=='GD':
            # Optimizer
            optimizer = tf.train.GradientDescentOptimizer(alpha).minimize(loss_total)
        elif optimizertype=='Adam':
            optimizer=tf.train.AdamOptimizer(alpha,beta1,beta2,epsilon).minimize(loss_total)
            
            
        predictions=logits

        
    elif loss == "CE":

        # Regular loss
        loss=tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=logits, labels=y))

        
        loss_total=loss+reg*reg_loss

        
        if optimizertype=='GD':
            optimizer = tf.train.GradientDescentOptimizer(alpha).minimize(loss_total)
        elif optimizertype=='Adam':
            optimizer = tf.train.AdamOptimizer(alpha,beta1,beta2,epsilon).minimize(loss_total)

        
        predictions=tf.sigmoid(logits)
        
    

    return weight,bias,x,y,reg,predictions,loss_total,optimizer
```

## 3.2 Implementing Stochastic Gradient Descent

```python
def SGD(alpha,beta1,beta2,epsilon,batch_size,epochs,loss,optimizertype,regSGD):
   
    weight,bias,x,y,reg,predictions,loss_total,optimizer=buildGraph(loss=loss,optimizertype=optimizertype,alpha=alpha,beta1=beta1,beta2=beta2,epsilon=epsilon)

    input_queue=tf.train.slice_input_producer([trainData,trainTarget],shuffle=True)
    batch_x,batch_y=tf.train.batch(input_queue, batch_size=batch_size)

    total_batch = int(np.floor(len(trainData)/batch_size))
    print('total number of batches=%i'%total_batch)

    # Initialize the variables (i.e. assign their default value)
    init = tf.global_variables_initializer()




    with tf.Session() as sess:

        sess.run(init)
        coord = tf.train.Coordinator()
        threads = tf.train.start_queue_runners(sess, coord)
        trainloss=[]
        validloss=[]
        testloss=[]
        trainacc=[]
        validacc=[]
        testacc=[]

        try:
            for epoch in range(epochs):
                avg_loss=0.
                for batch in range(total_batch):
                    batch_xs,batch_ys=sess.run([batch_x,batch_y])
                    _,cost=sess.run([optimizer,loss_total], feed_dict={x: batch_xs,y: batch_ys,reg:regSGD})
                    avg_loss+=cost/total_batch

                valid_loss=sess.run(loss_total,feed_dict={x: validData,y: validTarget,reg:regSGD})
                test_loss=sess.run(loss_total,feed_dict={x: testData,y: testTarget,reg:regSGD})
                trainloss.append(avg_loss)
                validloss.append(valid_loss)
                testloss.append(test_loss)
                
                train_acc=np.mean((predictions.eval({x: trainData,y: trainTarget})>=0.5)==trainTarget)*100
                valid_acc=np.mean((predictions.eval({x: validData,y: validTarget})>=0.5)==validTarget)*100
                test_acc=np.mean((predictions.eval({x: testData,y: testTarget})>=0.5)==testTarget)*100

                trainacc.append(train_acc)
                validacc.append(valid_acc)
                testacc.append(test_acc)
                
                if (epoch+1)%100==0 :
                    print("Epoch:", '%3d' % (epoch+1), "train loss=", "{:.4f}".format(avg_loss),
                          "valid loss=", "{:.4f}".format(valid_loss),"test loss=", "{:.4f}".format(test_loss),
                          "\n train accuracy=", "{:.4f}".format(train_acc),"valid accuracy=", "{:.4f}".format(valid_acc),
                          "test accuracy=", "{:.4f}".format(test_acc))                        
                        
            print("Optimization Finished!")
        except tf.errors.OutOfRangeError:
            print("done")
        finally:
            coord.request_stop()
        coord.join(threads)
    return trainloss,validloss,testloss,trainacc,validacc,testacc
```

```python
def plot_SGD(trainloss,validloss,testloss,trainacc,validacc,testacc,title,optimizertype,batch_size):
    epochs=range(100,700)
    plt.figure(figsize=(12,8))
    plt.plot(epochs,trainacc[100:],color='b',label='training accuracy')
    plt.plot(epochs,validacc[100:],color='r',label='validation accuracy')
    plt.plot(epochs,testacc[100:],color='g',label='test accuracy')
    plt.ylabel('accuracy',fontsize=16)
    plt.xlabel('iterations',fontsize=18)
    plt.title(title+' '+optimizertype+' '+" batchsize=%i"%batch_size,fontsize=20)
    plt.legend(loc='lower right')
    plt.savefig('accuracy plot')

    plt.figure(figsize=(12,8))
    plt.plot(epochs,trainloss[100:],color='b',label='training loss')
    plt.plot(epochs,validloss[100:],color='r',label='validation loss')
    plt.plot(epochs,testloss[100:],color='g',label='test loss')
    plt.ylabel('loss',fontsize=16)
    plt.xlabel('iterations',fontsize=18)
    plt.title(title+' '+optimizertype+' '+" batchsize=%i"%batch_size,fontsize=20)
    plt.legend(loc='upper right')
    plt.savefig('loss plot')
```

```python
def compare_plot(trainacc_mse,trainacc_ce):
    plt.figure(figsize=(12,8))
    plt.plot(trainacc_mse,color='b',label='linear regression')
    plt.plot(trainacc_ce,color='r',label='logistic regression')
    plt.ylabel('accuracy',fontsize=16)
    plt.xlabel('iterations',fontsize=18)
    plt.title('compare training accuracy',fontsize=20)
    plt.legend(loc='lower right')
    plt.savefig('SGD_compare.png')
```

```python
batch_size=500
alpha=0.001
epochs=700
beta1=0.9
beta2=0.999
epsilon=1e-08
loss='MSE'
optimizertype='GD'
title='linear regression'
regSGD=0
trainloss_mse,validloss_mse,testloss_mse,trainacc_mse,validacc_mse,testacc_mse=SGD(alpha,beta1,beta2,epsilon,batch_size,epochs,loss,optimizertype,regSGD,title)
plot_SGD(trainloss_mse,validloss_mse,testloss_mse,trainacc_mse,validacc_mse,testacc_mse,title,optimizertype,batch_size)
```
